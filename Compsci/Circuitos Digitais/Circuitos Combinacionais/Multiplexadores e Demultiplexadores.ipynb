# Multiplexador (MUX)

A função principal de um multiplexador é ser um seletor de sinais. Ele recebe vários sinais de entrada e escolhe um único para ser sua saída.

# Demultiplexador (DEMUX)

A função principal de um demultiplexador é replicar uma entrada para várias saídas. Ele recebe sinais de uma única entrada e replica para todas suas saídas.
