# Circuito Somador

O circuito somador, como o nome sugere, recebe a entrada de bits e produz binários em suas saídas, sendo alguns do resultado da soma e um de carry.

## Meio-Somador

O meio-somador recebe dois digitos binários de entrada e produz dois dígitos binários como saída, um de soma e outro de carry.

## Somador-Completo

O somador-completo recebe dois digitos binários de entrada e um de carry e produz uma saída de soma e um carry de saída.
