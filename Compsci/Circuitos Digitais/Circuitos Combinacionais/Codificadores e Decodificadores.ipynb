# Decodificadores

Um decodificador é um circuito que sinaliza a presença de uma combinação específica de bits em sua entradas, indicando a presença dessa sequência em sua saída.

# Codificadores

Um codificador é um circuito lógico que realiza a função inversa do decodificador, então recebe um digito decimal ou octal e converte em uma saída codificada (como binário, por exemplo).
