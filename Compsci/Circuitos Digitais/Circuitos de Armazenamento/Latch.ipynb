# Latches

Um latch é um tipo de dispositivo assíncrono biestável de armazenamento volátil, normalmente categorizado de forma diferente dos flip-flops. Embora similares aos flip-flops (dispositivos biestáveis que permanecem em um dos dois estados estáveis), a principal diferença entre os dois é o método usado para a mudança de estado de cada um.

O latch possui duas entradas, uma para SET e outra para RESET, e uma saída. A entrada SET faz com que a saída seja 1, a entrada RESET faz com que a saída seja 0. Caso ambas as entradas sejam acionadas, a saída se mantém ao que estava.
