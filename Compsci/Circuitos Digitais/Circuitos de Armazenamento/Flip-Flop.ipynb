# Flip-Flop

Flip-flops são dispositivos biestáveis síncronos de armazenamento volátil. Possuem três entradas: SET, RESET e TRIGGER. Possui uma saída Q.

- SET: Faz com que a saída seja 1.
- RESET: Faz com que a saída seja 0.
- TRIGGER: Mudança de estado que permite a alteração da saída.

## Saídas Flip-Flop Set-Reset
- SET=1, RESET=0 => Q = 1
- SET=0, RESET=1 => Q = 0
- SET=0, RESET=0 => Preserva o estado de Q
- SET=1, RESET=1 => Estado inválido

## Saídas Flip-Flop JK
- SET=1, RESET=0 => Q = 1
- SET=0, RESET=1 => Q = 0
- SET=0, RESET=0 => Preserva o estado de Q
- SET=1, RESET=1 => Inverte o estado anterior de Q (Toggle)
