# Busca Binária

A busca binária é um algoritmo de busca extremamente eficiente e simples usado para encontrar um valor específico dentro do vetor ordenado de tamanho $n$. Tem como base a estratégia de divisão e conquista, eliminando metade dos elementos a cada iteração.

Cada iteração do processo se baseia em três valores: _low_, _high_ e _mid_.

- Low: índice do primeiro elemento (geralmente 0)
- High: índice do último elemento (geralmente $n-1$)
- Mid: índice do elemento do meio ($\text{Mid} = \text{Low} + \frac{\text{High}-\text{Low}}{2}$)

Enquanto $\text{Low} \leq \text{High}$ as iterações continuam ocorrendo e toda vez são consideradas os seguintes casos:

1. Valor em Mid é igual ao alvo
  - Elemento encontrado, a busca termina e o índice de Mid é retornado.
2. Valor em Mid é maior que o alvo
  - Elemento está na metade esquerda do vetor.
  - A metade direita é descartada e $High = Mid - 1$.
3. Valor em Mid é menor que o alvo
  - Elemento está na metade direita do vetor.
  - A metade esquerda é descartada e $Low = Mid + 1$.

Se, em algum momento, $Low > High$, então o elemento procurado não está no vetor (geralmente retornando -1 num retorno inteiro ou _None_ no retorno do tipo _Option_).
