# Bubble Sort

É um algoritmo ineficiente, porém, fácil de entender. É baseado em varreduras feitas pelo array, analisando cada índice e seu subsequente. Caso o subsequente seja menor, um _swap_ é feito.

```c
#include <stdio.h>

void swap(int *a, int *b) {
  int aux = *a;
  *a = *b;
  *b = aux;
}

void bubble_sort(int *array, int n) {
  int swapped;

  for (int j=0; j<n-1; j++) {
    swapped = 0;
    for (int k=0; k<(n-j-1); k++) {
      if (array[k] > array[k+1]) {
        swap(&array[k], &array[k+1]);
        swapped = 1;
      }
    }

    if (!swapped)
      break;
  }
}

void print_array(int *array, int n) {
  for (int j=0; j<n; j++) printf("%d%s", array[j], j == n-1 ? "\n" : ", ");
}

int main() {
  int arr[] = {2, 5, 7, 3, 8};

  print_array(arr, 5);
  bubble_sort(arr, 5);
  print_array(arr, 5);

  return 0;
}
```
