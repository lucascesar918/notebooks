# Selection Sort (Ordenação por Seleção)

Na ordenação por seleção, o vetor é ordenado pegando seu menor valor e trocando de lugar com seu primeiro valor não ordenado, realizado sucessivamente até a completude da ordenação.

Sua eficiência é de:

- Melhor caso
  - Comparações: $O(n^2)$
  - Swaps: $O(1)$
- Caso médio
  - Comparações: $O(n^2)$
  - Swaps: $O(n)$
- Pior caso
  - Comparações: $O(n^2)$
  - Swaps: $O(n)$

É um algoritmo insatisfatório em termos de eficiência, por isso não é comumente utilizado em contextos que necessitem de alta performance.

## Exemplo de Selection Sort Crescente

```c
#include <stdio.h>

void swap(int *a, int *b);
void print_vector(int *vec, const size_t size);
void asc_selection_sort(int *vec, const size_t size);

int main() {
  const size_t size = 10;
  int vec[] = {5, 7, 2, 8, 9, 3, 4, 6, 5, 1};
  printf("Não ordenado: ");
  print_vector(vec, size);

  printf("Ordenado: ");
  asc_selection_sort(vec, size);
  print_vector(vec, size);

  return 0;
}

void swap(int *a, int *b) {
  int aux = *a;
  *a = *b;
  *b = aux;
}

void print_vector(int *vec, const size_t size) {
  for (int j=0; j<size; j++)
    printf("%d%s", vec[j], j==size-1 ? "" : ", ");
  putchar('\n');
}

void asc_selection_sort(int *vec, const size_t size) {
  int idx_low;

  for (int first_unsorted = 0; first_unsorted<size; first_unsorted++) {
    idx_low = first_unsorted;
    for (int j=first_unsorted; j<size; j++)
      if(vec[j] < vec[idx_low])
        idx_low = j;
      
    swap(&vec[idx_low], &vec[first_unsorted]);
  }
}
```

## Exemplo de Selection Sort Descrescente

```c
#include <stdio.h>

void swap(int *a, int *b);
void print_vector(int *vec, const size_t size);
void des_selection_sort(int *vec, const size_t size);

int main() {
  const size_t size = 10;
  int vec[] = {2, 5, 7, 2, 1, 6, 8, 4, 6, 3};

  printf("Não ordenado: ");
  print_vector(vec, 10);

  des_selection_sort(vec, 10);
  printf("Ordenado: ");
  print_vector(vec, 10);

  return 0;
}

void swap(int *a, int *b) {
  int aux = *a;
  *a = *b;
  *b = aux;
}

void print_vector(int *vec, const size_t size) {
  for (int i=0; i<size; i++)
    printf("%d%s", vec[i], i==size-1 ? "" : ", ");
  putchar('\n');
}

void des_selection_sort(int *vec, const size_t size) {
  int idx_high;

  for (int first_unsorted = 0; first_unsorted < size; first_unsorted++) {
    idx_high = first_unsorted;
    for (int j=first_unsorted; j<size; j++)
      if (vec[j] > vec[idx_high])
        idx_high = j;

    swap(&vec[idx_high], &vec[first_unsorted]);
  }
}
```
