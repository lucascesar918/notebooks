# Linked List

A linked list is a dinamic linear data structure. It is not expected for each node to be stored in a contiguously way (as it is usually heap-stored), though it still works.

A common practice for linked lists is having a head and a tail, which means mantaining a pointer to both first and last nodes.

## Singly Linked List

This variation contains two members:

- Data
- Pointer to the next node

```c
typedef struct {
  int data;
  Node *next;
} SinglyLLNode;
```

## Doubly Linked List

In this case there are three members:

- Data
- Pointer to the previous node
- Pointer to the next node

```c
typedef struct {
  int data;
  Node *previous;
  Node *next;
} DoublyLLNode;
```

## Circular Linked List

For this particular variation, there is no change regarding the members but the way the last node works. In a Circular Linked List, the last node points to the first node of the linked list.
