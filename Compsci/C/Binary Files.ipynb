# Binary Files

A file with raw bytes as content is a binary file. It is not human-readable as it is not made up of ASCII content.

All the major functions used for binary I/O are contained in ```stdio.h``` library.

## Binary Reading (rb)

The most common way to read binary files is to use ```fread```, it contains four parameters.

- **buffer**: pointer to the array where the objects are going to be stored.
- **size**: size of each object in bytes.
- **count**: the number of objects to be read.
- **stream**: pointer to input stream (e.g. pointer to a binary file).

The return value will be the number of bytes read, which will be less than **count** if there is an error or EOF.

If the stored content's size is not known, it can be read byte-per-byte using a **size** of 1 (byte) and a count of the **buffer** size.

## Binary Writing (wb)

The most common way to write binary files is to use ```fwrite```, it contains four parameters.

- **buffer**: pointer to the array where the content is stored.
- **size**: size of each object in bytes.
- **count**: the number of objects to be written
- **stream**: pointer to the output stream.

The return value will be the number of objects written, which will be less than **count** if there is an error.

## Handling File Pointer Position

The offset of a file pointer position can be obtained with ```ftell```, the file pointer position can be managed using ```fseek```.

### ```ftell```

This function takes a ```FILE*``` stream and returns a ```long``` value representing the file position indicator.

### ```fseek```

This function takes the following parameters:

- ```FILE* stream```: file stream to be used.
- ```long offset```: number of bytes to be shifted from origin.
- ```int origin```: position to which ```offset``` is added.
  - ```SEEK_SET```: start of file.
  - ```SEEK_END```: end of file.
  - ```SEEK_CUR```: current position of file.
