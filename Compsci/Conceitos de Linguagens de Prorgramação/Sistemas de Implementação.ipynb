# Implementações de Linguagens de Programação

As linguagens de programação podem ser implementadas por um dos três métodos:

1. Compilador: código-fonte traduzido para linguagem de máquina, gerando um binário para ser executado
2. Interpretador: cada comando é traduzido por vez para linguagem de máquina e então é executado comando por comando
3. Híbrido: meio-termo entre compilado e interpretado

## Compiladores

Num compilador existem vários processos subsequentes que seguem uma lógica para no final gerar um binário executável. Uma das primeiras fases do funcionamento de um compilador é a análise, dentre elas existem as seguintes frentes da análise:

- Léxica: captura as unidades léxicas (como identificadores, palavras-chave da linguagem, operadores e pontuação), comumente chamadas de _tokens_
- Sintática: 
- Semântica:

### Análise Léxica

Durante a análise léxica são capturadas as unidades léxicas (como identificadores, palavras-chave da linguagem, operadores e pontuação), também chamadas de _tokens_. Durante esse processo o compilador ignora os comentários, pois não há uso para os comentários durante a compilação.

### Análise Sintática

Após reunir as unidades léxicas, elas são organizadas em árvores sintáticas (_parse trees_). Muitas vezes as árvores sintáticas nem são utilizadas, pois as informações são utilizadas diretamente.

### Análise Semântica

O analisador semântico é parte do gerador de código intermediário, procurando erros (como os de tipo) que são difíceis ou impossíveis de detectar durante a análise sintática.

### Otimização

A otimização normalmente melhora os programas tornando-os menores, mais rápidos ou ambos ocorre no código intermediário. É uma parte opcional da compilação. Além disso, como há várias formas de otimização, são difíceis de serem feitas linguagem de máquina.

### Tradução

O gerador de código traduz o código intermediário (otimizado ou não) em um programa equivalente em linguagem de máquina. Para isso é utilizada uma tabela de símbolos, contendo uma base de dados com tipos, atributos de nomes, etc. Essa tabela de símbolos é alimentada a cada fase de análise (léxica, sintática e semântica) e também é utilizada na análise semântica.


### Linkedição

Apesar de ser possível executar a linguagem de máquina diretamente, geralmente ela precisa de algum outro código. A maioria dos programas também precisam de programas do sistema operacional, como entrada e saída. O compilador constrói chamadas para os programas do sistema, para isso esses programas de sistema devem ser encontrados e ligados ao programa do usuário. Fora os programas de sistema, geralmente os programas de usuário também usam bibliotecas, e é nessa fase onde essas bibliotecas sã́o ligadas a bibliotecas e a outros programas de usuário (se necessário).

## Interpretadores

A interpretação assume uma posição contrária a da compilação, onde programas são interpretados por outro programa (interpretador), sem tradução para código de máquina. Esse interpretador age como uma simulação em software de uma máquina, onde o ciclo de obtenção-execução trata, por vez, sentenças de um programa de alto nível ao invés de instruções de máquina. Essa simulação oferece uma máquina virtual para a linguagem.

Uma de suas grandes vantagens é facilidade de implementar muitas operações de depuração em código-fonte, pois todas as mensagens de erro em tempo de execução podem referenciar unidades de código-fonte. Então, caso um índice de vetor estiver fora da faixa permitida, uma mensagem de erro pode indicar no código onde exatamente onde está o erro e em qual vetor está o erro. Sua maior desvantagem geral é que geralmente o tempo de execução dessas linguagens são de 10 a 100 vezes mais lentas do que as linguagens compiladas, pois cada sentença é muito mais complexa e tem que ser decodificada em tempo de execução.

## Híbrido

Em alguns sistemas de implementação de linguagens os programas são traduzidos de linguagem de alto nível para uma linguagem intermediária para facilitar a interpretação. Esse método é mais rápido que a interpretação pura.

Um implementador pode fornecer tanto implementação compiladas quanto interpretadas para uma linguagem, usando um para desenvolver e depurar programas e o outro para lançar código com uma maior velocidade de execução.
