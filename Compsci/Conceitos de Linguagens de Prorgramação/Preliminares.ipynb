# Preliminares

Para avaliar linguagens de programação, usamos três critérios principais: legibilidade, facilidade de escrita e confiabilidade. São aspectos importantes não só para o uso dessas linguagens, mas também para o desenvolvimento dessas linguagens e de outras (que podem ou não ser baseadas nelas).

## Legibilidade

O critério de legibilidade avalia a afinidade que o usuário da linguagem tem ao ler e entender o código escrito nela. Em casos mais antigos, onde o foco das linguagens não era o usuário, a legibilidade era sacrificada pois o foco era a performance da linguagem em _hardwares_ limitados. Conforme a Lei de Moore foi afrouxando os requisitos técnicos, as linguagens foram tomando mais liberdade para fazer um código mais abstrato para a máquina e mais absorvível para pessoas.

A análise sobre a legibilidade de uma linguagem é importante principalmente pelos seguintes motivos: facilidade de desenvolvimento de código, facilidade de manutenção de código e se uma linguagem resolve um problema de forma natural.

- Linguagens pouco legíveis
    - COBOL
    - Assembly
    - Brainfuck
- Linguagens muito legíveis
    - Python
    - Rust
    - C#

Cada uma dessas liguagens se destaca em pouca/muita legibilidade em diferentes aspectos dentro do que é analisado como legibilidade.

### Simplicidade Geral

Uma linguagem com muitas construções básicas é sempre mais difícil de se aprender do que uma com poucas. Quando programadores precisam usar uma linguagem extensa, eles aprendem e usam um subconjunto dessa linguagem e ignoram os outros recursos. Isso se dá, por exemplo, pela multiplicidade de recursos. Quando feita de forma certa, pode simplificar a linguagem, como usar o operador + para soma entre inteiros e pra soma entre _floats_, reduzindo o número de operadores e simplificando a linguagem. Assim como usar o operador + para junção de vetores pode complicar a linguagem, pois não necessariamente é tão fácil de se ter o significado (significa a soma dos índices de cada vetor, a união dos vetores, uma operação lógica de OR entre vetores de bytes ou outro significado?).

Quando feita da forma certa, simplifica a linguagem, mas conforme o desenvolvimento da linguagem, necessariamente se esbarra numa necessidade que acaba aumentando o número de formas de se fazer a mesma coisa: aumentando o número de palavras-chave da linguagem, criando novas formas de se escrever a mesma coisa, dificultando a compreensão do código numa perspectiva geral, etc.

### Ortogonalidade

Resumidamente, a ortogonalidade é referente a combinações válidas da combinação de um conjunto relativamente pequeno de construções primitivas. Assim, quanto menos exceções dentro dessas combinações, mais ortogonal é a linguagem. Caso exista inteiros, vetores e ponteiros numa linguagem, um ponteiro poder apontar para vetores e para ponteiros significa um avanço na ortogonalidade da linguagem. Porém, caso hajam funções nessa linguagem, se um ponteiro não puder apontar para uma função, isso resulta numa menor ortogonalidade da linguagem. Um exemplo em C pode ser que funções não podem ser passadas como parâmetro, apenas ponteiros de função.

São notáveis em ortogonalidade as linguagens funcionais (como Lisp, Haskell, etc.), pois dado o pequeno conjunto de construções primitivas, suas combinações têm poucas exceções. Por questões de eficiência e performance linguagens funcionais acabam muitas vezes ficando de lado.

### Tipos de Dados

A presença de tipos ou mecanismos que possibilitem expressar tipos e estruturas de dados é um fator significativo para a legibilidade. Um exemplo da falta de legibilidade que a ausência de tipos de dados causam acontece em C, onde um tipo numérico (int) tem que ser utilizado para operações lógicas ao invés de um tipo booleano.

```c
// Menos compreensível, é estranho para quem ainda não está acostumado com a linguagem.
int logicaNumerico = 1;

// Muito mais legível, o que é lido é facilmente entendido
bool logicaBooleano = true;
```

### Projeto de Sintaxe

A sintaxe da linguagem reflete diretamente na sua legibilidade, especialmente em algumas escolhas como: se terá uso de chaves (e.g. C), begin/end (e.g. Pascal) ou indentação (e.g. Python) para blocos de instrução, se variáveis podem receber nome de palavras especiais como nome, como projetar cada sentença de forma que ela descreva (mesmo que minimamente) seu propósito e significado, etc.

## Facilidade de Escrita

A facilidade de escrita é denominada pelo quão fácil é desenvolver código para uma linguagem. A maioria das características que afetam a legibilidade afetam também a facilidade de escrita, já que o programador precisa reler o que já foi escrito para continuar desenvolvendo. Assim como na legibilidade, a facilidade de escrita deve ser avaliada dentro do escopo de problema que a linguagem busca resolver. Não é justo comparar a facilidade de escrita de duas linguagens com um problema dentro do domínio de somente uma delas.

### Simplicidade e Ortogonalidade

Se uma linguagem tem uma vasta quantidade de construções básicas, facilmente o programador pode acabar não conhecendo todas elas. Dessa forma se torna muito mais fácil de recursos indevidos serem usados ao invés de outros mais adequados, que seriam muito mais eficientes e/ou elegantes.

### Expressividade

Em geral, a expressividade de uma linguagem é obtida conforme ela pode ser escrita de uma forma mais concisa e elegante. Pode ser atingida usando operadores poderosos para realizar várias operações, diminuindo a verbosidade da linguagem e tornar coisas comuns da linguagem mais convenientes.

## Confiabilidade

Confíavel é aquela linguagem que está de acordo com suas especificações em todas as condições. Para isso são realizados testes, tratamentos e práticas num geral para garantir a confiabilidade da linguagem e dos programas feitos usando tal linguagem.

### Verificação de Tipos

A verificação de tipos pode ser feita durante a compilação (o ideal) ou durante a execução (que sempre será mais caro), garantindo que não haja erros de tipos. 

### Tratamento de Exceções

A capacidade de interceptar erros e mitigá-los em tempo de execução é um grande pilar da confiabilidade de uma linguagem.

### Apelidos

Apelidos ocorrem quando em um mesmo programa, uma célula de memória tem seu valor acessado por múltiplos nomes distintos. A maioria das linguagens permite algum tipo de "apelidagem", mesmo que seja com dois ponteiros apontando para a mesma célula de memória. É geralmente aceito que é uma funcionalidade perigosa em linguagens de programação, mas podem ser utilizados para compensar recursos para abstração de dados.

### Legibilidade e facilidade de escrita

Um problema resolvido de forma natural com uma linguagem que serve seu propósito dificilmente estará errado. Quando um problema é resolvido com estratégias artificiais, maior a probabilidade de estar errado. Além disso, a legibilidade afeta tanto o desenvolvimento do código quando sua manutenção, pois programas difíceis de ler também são difíceis de desenvolver e manutenir.

## Custo

O custo total de uma linguagem de programação é um produto de suas características. Também não é algo singular, tem diferentes faces que, no fim, representam o custo total.

1. Custo de treinamento: consiste em treinar programadores para usar a linguagem, uma função da simplicidade e ortogonalidade da linguagem. Linguagens mais poderosas não são necessariamente mais difíceis de aprender, mas normalmente são.

2. Custo de escrever: consiste na facilidade de desenvolver código na linguagem, o que aproxima mais do propósito da aplicação em particular e da facilidade de escrita.

3. Custo de compilação: consiste tanto no custo de processamento para rodar compiladores quanto no custo de desenvolver compiladores para a linguagem. Existe um trade-off entre confiabilidade e custo de compilação, onde a checagem de tipos em tempo de compilação acarreta num maior custo de compilação (por exemplo Rust, cujo compilador `rustc` é popularmente conhecido por ter extensos tempos de compilação mas que é extremamente confiável). Um caso notavel é Ada, onde a linguagem era muito extensa e desenvolver compiladores para a linguagem era um trabalho dispendioso e complexo.

4. Custo de execução: consiste no custo de processamento durante o tempo de execução do programa. Conforme as verificações de confiabilidade são delegadas para o tempo de execução, mais penosa é a execução do programa em termos de processamento. Um caso notável é que um programa que imprime números de 1 a 1 milhão em Python será muito mais lento em execução do que um similar em C.

5. Custo de implementação: consiste nos esforços necessários para portar para diferentes plataformas (como sistemas operacionais e arquiteturas). Esse foi um dos maiores motivos da grande aceitação da linguagem Java, pois bastava portar a JVM para cada plataforma.

6. Custo de confiabilidade: consiste no impacto que uma falha (que pode ser crítica) tem no mundo real. Um programa usado em uma usina nuclear, numa máquina de raio-X e numa cirurgia não podem falhar, caso contrário há grandes consequências negativas. Porém, também existem sistemas de simulação rápida onde diversos cálculos podem falhar em precisão que as consequências são aceitáveis.

7. Custo de manutenção: consiste na dificuldade para incluir correções e novas funcionalidades. O custo da manutenção envolve a legibilidade do código, afinal, geralmente, quem faz a manutenção do código não é quem o escreveu. Em softwares com tempos de vida longo, os custos de manutenção podem atingir até o quádruplo dos custos de desenvolvimento.

Otimizações podem ser incluídas para balancear os custos de compilação e execução, então código pode ser tanto gerado de forma menos custosa quanto executado de forma menos penosa.

## Influências no Projeto de Linguagens

Atualmente o padrão para a arquitetura de computadores são as máquinas de _Von Neumann_, então, de forma genérica, é esperado um _input_, que será processado entre a CPU e a memória (onde estão os dados e o programa) e direcionado a um _output_. Esse padrão tem forte influência em como as linguagens são projetadas, principalmente as de paradigma imperativo.

### Linguagens Imperativas

- Recursos principais:
    - Variáveis (uma abstração de célula de memória)
    - Sentenças de atribuição
    - Repetições iterativas
- Principais consequências:
    - Código altera o estado do programa
    - Iterações são rápidas pois instruções são armazenadas sequencialmente, sendo necessário apenas um desvio de fluxo, e não a recursão (embora em alguns casos seja mais elegante)
    - Execução de programa baseada em um ciclo de obtenção (das instruções e dados) e execução (das instruções com os dados)

### Linguagens Funcionais

- Recursos principais:
    - Foco em cálculo lambda
    - Funções são tratadas como valores comuns, podendo ser argumentos, retornadas e armazenadas em estruturas de dados
    - Imutabilidade, onde um valor sempre será atribuído a um nome, o que existe é a criação de novos dados a partir dos preexistentes
    - Recursão, pois não existem repetições iterativas
    - Foco em não alterar o estado do programa
- Principais consequências:
    - Funções são puras, então não há valores globais que alterem o resultado de uma função (já que evita alterar o estado do programa)
    - Referências são transparentes, então funções podem ser substituídas pelo seu valor sem alterar o funcionamento do programa, facilitando testes e provas de correção
    - Geralmente funções não são calculadas até que seja necessário, então existe como manipular estruturas de dados infinitas e otimizar a performance em certos cenários
    - Foco na descrição das transformações de dados através da composição de funções, resultando num código mais abstrato e conciso
    - Embora a recursão exija mais da pilha, é muito mais próxima da lógica-matemática e é muito otimizada por técnicas modernas
