# Teoremas de Bayes e da Probabilidade Total

Para facilitar, é bom pensar como uma relação de causas e efeitos.

- **Causas:** diferentes origens de algo (ex: Máquina A, Máquina B, Máquina C)
- **Efeito:** o que acontece no final (ex: sair uma peça com defeito)

## Partição do Espaço Amostral

Para usar os teoremas, as causas precisam ser partições perfeitas, o que significa que:

1. **São mutuamente exclusivas:** duas causas não são coexistentes pro efeito (ex: uma peça não pode ser fabricada pelas máquinas A e B ao mesmo tempo)
2. **São exaustivas (cobrem 100%):** a soma das causas tem que resultar em 100%, não existe uma causa oculta (ex: a soma da produção das máquinas A,B e C cobrem 100% da produção, não existe uma máquina D oculta)

## Teorema da Probabilidade Total

Imagine que o diretor da fábrica pergnte a probabilidade de pegar uma peça defeituosa qualquer no fim do dia. Não dá pra simplesmente somar as taxas de defeito das três máquinas, pois cada uma produz uma quantidade diferente de peças. O **Teorema da Probabilidade Total** diz que você deve calcular o defeito de cada ramo e somar.

$$
    P(B) = \sum_i{P(A_i) \cdot P(B|A_i)}
$$

- **$P(B)$:** Probabilidade de uma peça defeituosa vir da máquina B
- **$P(A_i)$:** Probabilidade de uma das causas, é a fatia que cada um tem do universo total
- **$P(B|A_i)$:** Taxa de peças estragadas que existem dentro da fatia $P(A_i)$.
- **$P(A_i) \cdot P(B|A_i)$:** Quantidade de peças defeituosas que a máquina produziu.

Isso significa que: **Probabilidade Total de Defeito = (Chance de ser uma peça da Máq. A $\times$ chance de defeito da Máq. A) + (Chance de ser uma peça da Máq. B $\times$ chance de defeito da Máq. B) + (Chance de ser uma peça da Máq. C $\times$ chance de defeito da Máq. C)**.
