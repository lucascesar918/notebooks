# Princípios Fundamentais da Contagem
> Os Princípios Fundamentais da Contagem, ou PFC, são alguns princípios para a resolução de problemas envolvendo contagens numerosas e de difícil contagem.

## Princípios Aditivo e Multiplicativo

Caso seja uma etapa com o conectivo lógico **e**, devem ser somadas as contagens.

> Ex: probabilidade de uma pessoa gostar de futebol **e** de basquete.

Caso seja uma etapa com o conectivo lógico **ou**, devem ser multiplicadas as contagens.

> Ex: probabilidade de uma pessoa gostar de futebol **ou** de basquete.

## Princípio da Preferência

Começar sempre pela contagem do caso mais exclusivo.

# Agrupamentos

Quando a natureza do elemento não está sendo alterada e somente a organização desses elementos é alterada, está ocorrendo um agrupamento. Considere o número de elementos $n$ e o número $p$ de elementos no resultado.



```python
from graphviz import Digraph

# Configuração do gráfico direcionado (Top-Down)
dot = Digraph(comment='Análise Combinatória')
dot.attr(rankdir='TD', size='6,6')
dot.attr('node', fontname='Helvetica', fontsize='12')

# Estilo para os nós de decisão (Losangos)
dot.attr('node', shape='diamond', style='filled', fillcolor='#e6f2ff', color='#0066cc')
dot.node('A', 'A ordem importa?')
dot.node('B', 'p = n')

# Estilo para os nós de resultado (Retângulos arredondados)
dot.attr('node', shape='box', style='filled, rounded', fillcolor='#f2f2f2', color='#666666')
dot.node('C', 'Combinação')
dot.node('D', 'Permutação')
dot.node('E', 'Arranjo')

# Definição das arestas com rótulos
dot.edge('A', 'B', label=' Sim')
dot.edge('A', 'C', label=' Não')
dot.edge('B', 'D', label=' Sim')
dot.edge('B', 'E', label=' Não')

# A chamada do objeto renderiza o SVG no output da célula
dot
```

## Arranjo

> Arranjar elementos significa que a ordem dos elementos altera o resultado, também considerando um número maior de elementos disponíveis do que o número de elementos no resultado ($p \neq n$).

Por exemplo, existem três vagas a serem ocupadas por cinco pessoas. O número de vagas não comporta as cinco pessoas, então precisaremos arranjá-las. Sendo o número de vagas $p$, o número de pessoas $n$ e o número de arranjos dessas vagas $A_{p,n}$, temos que:

$$
    A_{p,n} = \frac{n!}{(n-p)!}
$$

## Combinação

> Nas combinações a ordem dos elementos não influencia o resultado. Mas nem todos os elementos disponíveis são contemplados no resultado ($p \neq n$).

$$
    C_{p,n} = \frac{n!}{p! \cdot (n-p)!}
$$

## Permutação

> Permutação, ou permutar, significa trocar de ordem. Logo, análises de permutaçao são úteis quando o número de elementos disponíveis é igual ao número de elementos no resultado ($p=n$).

### Permutação Simples

Numa permutação simples, é suposto que todos os elementos analisados não serão repetidos e preencherão os $n$. Logo, é fácil contá-los por $n!$ ou $p!$.

### Permutação com Repetição

Quando elementos são repetidos, teremos que anular casos onde há uma só ordem específica sendo contada duas vezes. Para anular esses casos, temos que usar:

$$
    P_n = \frac{n!}{r_1! \cdot r_2! \cdot r_3! \cdot ...}
$$

Nesse caso a permutação dos $p$ elementos em $n$ é igual ao fatorial de $n$ sobre o produto entre quantas repetições tem cada elemento.

Ex: anagramas de **ARARA**

- $n = 5$, pois há cinco espaços para as letras
- $r_1 = 3$, pois a letra **A** será repetida 3 vezes
- $r_2 = 2$, pois a letra **R** será repetida 2 vezes

Então:

$$
    P_x = \frac{5!}{3! \cdot 2!} \implies P_x = 10
$$
