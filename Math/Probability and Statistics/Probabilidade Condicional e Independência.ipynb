# Probabilidade Condicional e Independência

Quando tratamos da probabilidade de ocorrer um evento após um outro evento já ter ocorrido, é útil analisarmos com a ótica da probabilidade condicional. Essa informação sobre uma condição que já ocorreu altera o espaço amostral, então o universo de possibilidades é reduzido.

## Frequência Conjunta

Sejam $A$ e $B$ dois eventos, a probabilidade de ocorrer $A$ dada a ocorência de $B$ é denotada por $P(A|B)$ e calculada pela fórmula seguinte:

$$
    P(A|B) = \frac{P(A \cap B)}{P(B)} \text{, se } P(B) > 0
$$

## Teorema da Multiplicação

Isolando $P(A \cap B)$ na fórmula da probabilidade condicional, multiplicando $P(B)$, é obtida a regra do produto:

$$
    P(A \cap B) = P(A|B) \cdot P(B)
$$

Na prática, isso significa que a chance de dois eventos ($A$ e $B$) ocorrerem simultaneamente é a mesma do primeiro acontecer multiplicada pela do segundo acontecer, dado que o primeiro já ocorreu.

## Independência de Eventos

Dois eventos são considerados independentes se a ocorrência de um não afeta a probabilidade da ocorrência do outro.

$$
    P(A|B) = P(A)
    P(B|A) = P(B)

    \text{Como consequência:}

    P(A \cap B) = P(A) \cdot P(B) \text{se, e somente se, A e B forem eventos independentes}
$$

Atenção para não confundir eventos independentes com eventos disjuntos (mutuamente exclusivos).

- Eventos disjuntos: não podem ocorrer ao mesmo tempo (P(A \cap B) = 0).
- Eventos independentes: podem ocorrer ao mesmo tempo, mas a ocorrência de um não altera a chance do outro ocorrer.
