# Definições Formais de Probabilidade

Para uma abordagem adequada da probabilidade de eventos, a estatística separa a probabilidade em três definições formais:

- Equiprovável (Clássica)
- Frequentista
- Axiomática de Kolmogorov

## Definição Equiprovável de Probabilidade

Útil em espaços amostrais finitos onde todos os resultados têm a mesma chance de ocorrer. A probabilidade, nesse caso, é a razão entre o número de evewntos favoráveis e o espaço amostral total:

$$
    P(A) = \frac{n(A)}{n(\Omega)}
$$

Se o espaço $\Omega$ tem $n$ elementos, a probabilidade de cada ponto amostral é $\frac{1}{n}$.

## Definição Frequentista

É uma definição empírica, então totalmente baseada na repetição de experimentos. A probabilidade de um evento $A$ é o limite da frequência relativa de $A$ quando o número de repetições $n$ tende ao infinito: $P(A) = \lim{n \to \infty} \frac{n_A}{n}$

## Definição Axiomática de Kolmogorov

É uma definição matemática rigorosa que define a probabilidade como uma função $P$ que mapeia um evento para um número real no intervalo $[0, 1]$

### Axiomas

Toda a probabilidade é sustentada por três axiomas definidos por Kolmogorov e se algo violar isso, está errado.

- **Axioma 1:** A probabilidade de $\Omega$ é sempre $1$. $P(\Omega) = 1$
- **Axioma 2:** A probabilidade de qualquer evento A é não negativa. $P(A) \geq 0$
- **Axioma 3:** Se há dois eventos mutuamente exclusivos ($P(A) \cap P(B) = \empty$), a probabilidade da união deles é a soma das suas probabilidades individuais.

### Propriedades e Regras Operacionais

- **Probabilidade do Evento Complementar:** A probabilidade de um evento $A$ não ocorrer, é $P(A') = 1 - P(A)$
- **Probabilidade do Conjunto Vazio:** Um evento impoossível tem probabilidade nula $P(\empty) = 0$.
- **Regra da Adição:** Se há dois eventos não-disjuntos (podem ocorrer simultaneamente), sua probabilidade se deve pela soma da probabilidade individual de cada subtraindo a intersecção: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$
